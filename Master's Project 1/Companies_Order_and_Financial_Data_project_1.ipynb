### Dropping unwanted columns

```python
dataFrame = dataFrame.drop(['index','Freight','TCS','CGST %','SGST %','IGST %','Mat No','Sale_Order','Sale_It','Cust_PO_Dt',
                            'CancBillDc','Can.Ex.In','Sal Off','CoCode','Region','RegCode'], axis= 1)
```

We are dropping these columns as they do not have any significant value during visualization of the data

```python
dataFrame
```

After Dropping the unwanted columns we can see that there are still few null values present in the dataset

```python
dataFrame.isnull().sum()
```

```python
dataFrame['Material_Descr'].fillna(dataFrame.Material_Descr.mode()[0],inplace=True)
dataFrame['Executive_Name'].fillna(dataFrame.Executive_Name.mode()[0],inplace=True)
dataFrame['Service_Executive'].fillna(dataFrame.Service_Executive.mode()[0],inplace=True)
dataFrame.isnull().sum()
```

We can not drop those null values as they are useful in the data visualisation. Dropping them may bias the dataset. So, we fill the null values present in  Material_Descr, Executive_Name, Service_Executive with the mode values. 

## Data Visualization

### Graph Representing The Material Used

```python
dataFrame.Material_Descr.value_counts().plot(kind='pie',figsize=(10,25))
plt.title("Material User")
```

In the above pie chart we can see that the most used material in the company is "Carton- Paperboard"

### Visualising the Gross Amount of Each Day in the month of September

```python
dataFrame.groupby(['Bill_Date'],sort=True)['Gross_Amnt'].sum().plot(kind='bar',figsize=(20,10))
plt.xlabel("Bill Date")
plt.ylabel("Gross Amount")
plt.title("Bill Date vs Gross Amount")
```

### Visualising the Quantity Billed for Each Day in the month of September

```python
dataFrame.groupby(['Bill_Date'],sort=True)['Billed_qty'].sum().plot(kind='bar',figsize=(20,10))
plt.xlabel("Bill Date")
plt.ylabel("Billed Quantity")
plt.title("Bill Date vs Billed Quantity")
```

```python
print(dataFrame.groupby(['Bill_Date'],sort=True)['Gross_Amnt'].sum().sort_values(ascending=False).head(5))
print(dataFrame.groupby(['Bill_Date'],sort=True)['Billed_qty'].sum().sort_values(ascending=False).head(5))
```

From the above graphs we can see that the Gross Amount is highest on 29th September 2022. But the Billed quantity is highest on 26th September 2022.

### Visualizing the sum of tunover gained by each Executive Employee (Those who get the orders to the company)

```python
dataFrame.groupby(['Executive_Name'],sort=True)['Bal_Trnovr'].sum().plot(kind='bar',figsize=(18,8))
plt.xlabel("Executive Name")
plt.ylabel("Turnover")
plt.title("Executive Name vs Turnover Gained")
```

### Visualizing the sum of Billed Quantity by each Executive Employee (Those how get the orders to the company)

```python
dataFrame.groupby(['Executive_Name'],sort=True)['Billed_qty'].sum().plot(kind='bar',figsize=(18,8))
plt.xlabel("Executive Name")
plt.ylabel("Billed Quantity")
plt.title("Executive Name vs Billed Quantity")
```

```python
print(dataFrame.groupby(['Executive_Name'],sort=True)['Bal_Trnovr'].sum().sort_values(ascending=False).head(5))
print(dataFrame.groupby(['Executive_Name'],sort=True)['Billed_qty'].sum().sort_values(ascending=False).head(5))
```

The above two graphs represnt who is getting more order to the company and who is getting the significant profits with it. Using this data we can consider the one who is getting more order for the salary hikes.
